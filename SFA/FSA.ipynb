```python
# Split the string into lines
import json
lines = output.split('\n')

# Create a dictionary
ds = {}

data_dict = {}
for line in lines:
    key_value = line.split(' = ')
    key = key_value[0].split('. ')[1]
    value = key_value[1].strip('\"').split(", ")
    data_dict[key] = value
ds[word] = data_dict
json_data = json.dumps(ds, ensure_ascii=False, indent=4)
print(json_data)
```
